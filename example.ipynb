```python
import cv2
import support  #マジックコマンド cv_img を使用する
```

```python
img = cv2.imread("lena.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
import glob
# 検出器用xmlの保存場所
haarcascades = "{}share/OpenCV/haarcascades/haarcascade_frontalface_default.xml".format(glob.glob('/opt/conda/pkgs/opencv-*/')[0])
face_cascade = cv2.CascadeClassifier(haarcascades)
```

```python
for (x,y,w,h) in face_cascade.detectMultiScale(img_gray):
    cv2.rectangle(img, (x,y), (x+w, y+h), (0, 255, 0), 2)
```
